```python
!pip install edge-tts
import edge_tts
import asyncio
import IPython.display as ipd

```

```python
async def text_to_speech(text, file_name, voice="es-ES-AlvaroNeural", rate="+0%"):
    """Convierte texto en audio y lo guarda en un archivo."""
    tts = edge_tts.Communicate(text, voice=voice, rate=rate)
    await tts.save(file_name)
    print(f"✅ Audio guardado: {file_name}")
```

```python


async def main():
    # Título
    titulo = "Este es el título del hilo de Reddit."
    # Cuerpos
    cuerpos = [
        "Esta es la voz del cuerpo 1 del hilo.",
        "Esta es la voz del cuerpo 2 del hilo.",
        "Esta es la voz del cuerpo 3 del hilo.",
        "Esta es la voz del cuerpo 4 del hilo.",
        "Esta es la voz del cuerpo 5 del hilo.",
        "Esta es la voz del cuerpo 6 del hilo."
    ]

    # Voces
    voice_titulo = "es-CR-JuanNeural"  # Voz para el título
    voice_cuerpo = "es-CR-JuanNeural"  # Voz para los cuerpos
    rate = "+10%"  # Ajuste de velocidad

    # Lista de tareas asincrónicas
    tareas = [text_to_speech(titulo, "titulo.mp3", voice_titulo, rate)]

    # Agregar tareas para los 6 cuerpos
    for i, cuerpo in enumerate(cuerpos, start=1):
        tareas.append(text_to_speech(cuerpo, f"cuerpo_{i}.mp3", voice_cuerpo, rate))

    # Ejecutar todas las tareas en paralelo
    await asyncio.gather(*tareas)

    # Reproducir el audio del título como ejemplo
    #return ipd.Audio("titulo.mp3")
    return ipd.Audio("cuerpo_1.mp3")

# Ejecutar la función principal
await main()
```
